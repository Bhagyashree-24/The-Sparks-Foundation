<b>Load SampleSuperstore dataset

```r
SampleSuperstore <- read.csv("D:/Bhagyashree/The Sparks Foundation/SampleSuperstore.csv")
head(SampleSuperstore)
```

<b> Data Prepration

```r
# Checking if any duplicate rows present in the datatset
duplicated(SampleSuperstore)
```

```r
# In the dataset there are 17 duplicate row present in the dataset
which(duplicated(SampleSuperstore))
```

```r
# Removing 17 duplicate rows 
New_data= SampleSuperstore[-which(duplicated(SampleSuperstore)),]
(New_data)
 str(New_data)
```

```r
summary(New_data)
```

```r
# Checking the missing values
sum(is.na(New_data))
```

```r
# Removing unwanted columns or features and converting some columns into factors
New_data <- New_data[, !(names(New_data) %in% c("Country","Postal.Code"))]
New_data$Ship.Mode=as.factor(New_data$Ship.Mode)
New_data$Segment=as.factor(New_data$Segment)
New_data$Region=as.factor(New_data$Region)
New_data$Category=as.factor(New_data$Category)
New_data

```

```r
summary(New_data)
```

```r
# Importing library for plotting
library(ggplot2)
```

```r
# Category wise product distribution

ggplot(New_data, aes(Sub.Category, fill = factor(Category))) +
     geom_bar() + geom_text(stat = "count", aes(label = ..count..), hjust = 1) +
     coord_flip() + scale_fill_discrete(
         name = "Category"
   ) +labs(title = "Category wise Product distribution")

```

```r

lbls <- c("Furniture","Office Supplies", "Technology") 
 pct = round(table(SampleSuperstore$Category)/sum(table(SampleSuperstore$Category))*100)
 lbls = paste(lbls, pct) # add percent to labels
 lbls = paste(lbls,"%",sep="") # ad % to labels
           pie(table(SampleSuperstore$Category),labels = lbls, col=rainbow(length(lbls)), 
               main="Pie Chart of Category")
```

<b>Observation: Maximum qunatity is of office supplies category 

```r
# Region wise category Distribution

lbls <- c("Central","East", "South", "West") 
 pct = round(table(New_data$Region)/sum(table(New_data$Region))*100)
 lbls = paste(lbls, pct) # add percent to labels
 lbls = paste(lbls,"%",sep="") # ad % to labels
           pie(table(New_data$Region),labels = lbls, col=rainbow(length(lbls)), 
               main="Pie Chart of Region")
```

<b>Observation: Maximum category distributed in West region

```r


# Category wise sales in different region


ggplot(data = New_data, aes(
     x = Region,
     y = Sales,
     fill = factor(Category)
 )) +
     geom_boxplot(position = position_dodge()) +
     labs(
         title = "sales of Category ",
         subtitle = "In differnt Region",
         x = "Region",
         y = "Sales(count)"
     )
```

<b> Technology category has highest sales in all the Region

```r
# Region wise total sales

tapply(New_data$Sales,New_data$Region,sum,na.rm=TRUE)->Reg_sales
Reg_sales
barplot(Reg_sales,xlab="Region",ylab="Sales(count)",col="Orange")
```

<b>Observation: In West region highest sales are done

```r
# Category wise total sales
tapply(New_data$Sales,New_data$Category,sum,na.rm=TRUE)->cat_sales
cat_sales
barplot(cat_sales,xlab="Category",ylab="Sales(count)",col="Blue")
```

<b>Observation: Technology category has highest sales where office supplies category has lower sales but quantity is more

```r
# Sql queries in R using sqldf library
library(sqldf)
```

```r
# Checking State wise Total Sales,Discount,Profit,Quantity 

spd=sqldf("select distinct (State),sum(Sales) as Sales,sum(Discount) as Discount,sum(Profit) as Profit,sum(Quantity)as Qty
 from New_data group by State order by Sales desc")
 spd
```

<b>Observation: Max Sales are done in California state where as in Nort dakota min sales are done  

```r
# City wise Total Sales,Discount,Profit,Quantity 

spd_city=sqldf("select distinct (City),sum(Sales) as Sales,sum(Discount) as Discount,sum(Profit) as Profit,sum(Quantity) as Qty
from New_data group by City order by Sales desc")
spd_city
```

<b>Observation: Max sales are done in New york city where as in Abliene city min sales are done
